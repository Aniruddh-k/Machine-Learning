```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/workspaces/Machine-Learning/EDA/Titanic-Dataset.csv')
```

```python
df.head()
```

**Bivariate Analysis**

*Numerical - Numerical*

```python
sns.scatterplot(x=df['Age'],y=df['Fare'])
```

*Numerical - Categorical*

```python
#BarPlot
sns.barplot(x=df['Survived'],y=df['Age'])
```

```python
sns.barplot(x=df['Pclass'],y=df['Age'])
```

```python
#Boxplot
sns.boxplot(x=df['Pclass'],y=df['Age'])
```

```python
#distplot
sns.distplot(df[df['Survived'] == 0]['Age'],hist=False)
sns.distplot(df[df['Survived'] == 1]['Age'],hist=False)
```

*Categorical - Categorical*

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```
